### BODY TYPE PREDICTION

```python
# Let us import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)
```

```python
# Call the json object into a dataframe using pandas
df1 = pd.read_json('measurements.json')
df1
```

```python
df1.columns
```

```python
# Drop unnecessary columns
df1.drop(columns=['arm_length_cm', 'front_build_cm',
                  'hips_circumference_cm', 'leg_length_cm',
                  'neck_circumference_cm', 'neck_waist_length_front_cm',
                  'shoulder_width_cm', 'thigh_circumference_cm',
                  'under_chest_circumference_cm', 'weight',
                  'age', 'race', 'profession'],
        axis= 1,
        inplace = True)
```

```python
df1.shape
```

```python
df1
```

```python
df1.rename(columns={'chest_circumference_cm': 'bust',
                   'pelvis_circumference_cm': 'hips',
                   'waist_circumference_cm': 'waist'},
        inplace = True)  # rename the columns
```

```python
df1
```

```python
df1.info()
```

```python
# Remove trailing characters in columns that should be numerical and convert columns to numerical 
cols_to_clean = ['bust', 'hips', 'waist']

for col in cols_to_clean:
    df1[col] = df1[col].astype(str).str.extract(r'(\d+\.?\d*)')  # keep only numbers
    df1[col] = pd.to_numeric(df1[col], errors= 'ignore')  # convert back to float
```

```python
df1
```

```python
df1.info()
```

```python
# Convert our measurements from cm to inches
cm_to_inch = 0.3937  # unit of conversation
for col in cols_to_clean:
    if df1[col].dtype == int or float:
        df1[col] = df1[col]*0.3937
```

```python
# Convert the columns to integers
for col in df1.columns:
    if df1[col].dtype == float:
        df1[col] = df1[col].round().astype(int)
```

```python
df1['height'] = (df1['height']*0.3937).round().astype(int)
```

```python
df1
```

```python
df1.to_csv('data1.csv', index=False)
```
